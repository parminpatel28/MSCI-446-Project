```python
data_exploration_df = pd.read_csv('csvs/cumulative_averages.csv')
data_exploration_df
```

```python

df_season_2018 = data_exploration_df[data_exploration_df['season'] == 2018]


team0 = df_season_2018['team0']
team1 = df_season_2018['team1']

date = data_exploration_df['date']

date = pd.to_datetime(date)

positions = 1
bar_width = 0.35

# Plotting TOV% for team0
plt.bar([p - bar_width/2 for p in positions], 0.8, width=bar_width, label='Team 0 TOV%')

# Plotting TOV% for team1
plt.bar([p + bar_width/2 for p in positions], 0.1, width=bar_width, label='Team 1 TOV%')


plt.title('TOV% Over Time for Team 0 and Team 1')
plt.xlabel('Date')
plt.ylabel('TOV%')

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```

```python


heatmap_data = df_season_2018[['tov%_cumulative_team0', 'tov%_cumulative_team1']].transpose()
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Heatmap of TOV% for Team 0 and Team 1 in Season 2018')
plt.ylabel('Team')
plt.xlabel('Game Number')
plt.yticks(rotation=0)  # Rotate y-axis labels to horizontal

plt.show()
```

```python
sns.jointplot(data=df_season_2018, x='tov%_cumulative_team0', y='tov%_cumulative_team1', kind="kde", fill=True, space=0)

plt.suptitle('KDE Joint Plot of TOV% for Team 0 and Team 1 in Season 2018', y=1.02)
plt.xlabel('TOV% Cumulative Team 0')
plt.ylabel('TOV% Cumulative Team 1')
```

```python


# Mapping the 'team1_winner' to a color
df_season_2018['color'] = df_season_2018['team1_winner'].map({0: 'blue', 1: 'red'})


for color in ['blue', 'red']:
    subset = df_season_2018[df_season_2018['color'] == color]
    plt.scatter(subset['tov%_cumulative_team0'], subset['tov%_cumulative_team1'], color=color, label=f"Winner: {'Team 0' if color == 'blue' else 'Team 1'}")



plt.title('Scatter Plot of TOV% for Team 0 vs. Team 1 in Season 2018')
plt.xlabel('TOV% Cumulative Team 0')
plt.ylabel('TOV% Cumulative Team 1')

# Adding a reference line to indicate where TOV% is equal for both teams
# plt.axline((0, 0), slope=1, color='red', linestyle='--', label='Equal TOV% Line')

plt.legend()
plt.grid(True)
plt.show()
```

```python
avg_tov_team1_wins = data_exploration_df[data_exploration_df['team1_winner'] == 1]['tov%_cumulative_team1'].mean()

# Calculating average TOV% for Team 1 losses (which means Team 0 wins)
avg_tov_team1_losses = data_exploration_df[data_exploration_df['team1_winner'] == 0]['tov%_cumulative_team1'].mean()

# data_exploration_df.head(10)
print(f"Average tov for Team 1 wins: {avg_tov_team1_wins}")
print(f"Average tov for Team 1 losses: {avg_tov_team1_losses}")

# Data for plotting
tov_data = {'Outcome': ['Wins', 'Losses'], 'Average TOV%': [avg_tov_team1_wins, avg_tov_team1_losses]}

df_tov_outcome = pd.DataFrame(tov_data)

# Plotting
df_tov_outcome.plot(x='Outcome', y='Average TOV%', kind='bar', legend=False, color=['green', 'red'], figsize=(8, 6))
plt.title('Average TOV% for Team 1 Wins vs. Losses')
plt.xlabel('Outcome for Team 1')
plt.ylabel('Average TOV%')
plt.xticks(rotation=0)
plt.grid(axis='y')

plt.show()
```

```python
tov_percentages = pd.concat([
    pd.DataFrame({'TOV%': data_exploration_df['tov%_cumulative_team0'], 'Outcome': data_exploration_df['team1_winner'].replace({1: 'Loss', 0: 'Win'})}),
    pd.DataFrame({'TOV%': data_exploration_df['tov%_cumulative_team1'], 'Outcome': data_exploration_df['team1_winner'].replace({1: 'Win', 0: 'Loss'})})
])

# Create box plots
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='TOV%', data=tov_percentages)
plt.title('Distribution of TOV% by Game Outcome')
plt.ylabel('TOV%')
plt.xlabel('Game Outcome')
```

```python
plt.figure(figsize=(16, 10))
data = data_exploration_df
# List of statistics to plot
stats = ['efg%_cumulative_team1', 'tov%_cumulative_team1', 'orb%_cumulative_team1', 'ft/fga_cumulative_team1']
titles = ['Effective Field Goal Percentage (eFG%)', 'Turnover Percentage (TOV%)',
          'Offensive Rebound Percentage (ORB%)', 'Free Throw Rate (FT/FGA)']

for i, stat in enumerate(stats):
    plt.subplot(2, 2, i+1)
    sns.kdeplot(data[data['team1_winner'] == 1][stat], label='Wins', shade=True)
    sns.kdeplot(data[data['team1_winner'] == 0][stat], label='Losses', shade=True)
    plt.title(f'Density Plot of {titles[i]} (Wins vs Losses)')
    plt.xlabel(titles[i].split(' ')[0])
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Setting up the figure for box plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
stats = ['efg%_cumulative_team1', 'tov%_cumulative_team1', 'orb%_cumulative_team1', 'ft/fga_cumulative_team1']
titles = ['Effective Field Goal Percentage (eFG%)', 'Turnover Percentage (TOV%)',
          'Offensive Rebound Percentage (ORB%)', 'Free Throw Rate (FT/FGA)']

for ax, stat, title in zip(axes.flatten(), stats, titles):
    sns.boxplot(x='team1_winner', y=stat, data=data, ax=ax)
    ax.set_title(title)
    ax.set_xlabel('Game Outcome (0=Loss, 1=Win)')
    ax.set_ylabel(stat)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
stats = ['efg%_cumulative_team1', 'tov%_cumulative_team1', 
         'orb%_cumulative_team1', 'ft/fga_cumulative_team1']
titles = ['Effective Field Goal Percentage (eFG%)', 'Turnover Percentage (TOV%)',
          'Offensive Rebound Percentage (ORB%)', 'Free Throw Rate (FT/FGA)']

for ax, stat, title in zip(axes.flatten(), stats, titles):
    # Selecting data for wins and losses
    win_data = data[data['team1_winner'] == 1][stat]
    loss_data = data[data['team1_winner'] == 0][stat]
    
    # Plotting histograms
    ax.hist(win_data, bins=20, alpha=0.5, label='Wins')
    ax.hist(loss_data, bins=20, alpha=0.5, label='Losses')
    ax.set_title(title)
    ax.set_xlabel(stat)
    ax.set_ylabel('Frequency')
    ax.legend()

plt.tight_layout()
plt.show()
```
